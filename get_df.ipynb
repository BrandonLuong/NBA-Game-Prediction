```python
import pandas as pd
```

```python
games_raw = pd.read_csv('games_stats.csv')
games_raw
```

```python
games_raw.describe()
```

```python
games_raw = games_raw.drop(columns = ['Unnamed: 0', 'Season', '+/-', 'PF', ])
games_raw
```

```python
games_raw.columns
```

Separate home and away games 

```python
away_teams = games_raw.loc[games_raw['Match\xa0Up'].str.contains('@')]
away_teams = away_teams.reset_index(drop=True)
home_teams = games_raw.loc[games_raw['Match\xa0Up'].str.contains('vs.')]
home_teams = home_teams.reset_index(drop=True)
home_teams.rename(columns= lambda x: 'Home ' + x, inplace=True)
away_teams.rename(columns= lambda x: 'Away ' + x, inplace=True)
list(away_teams.columns)
```

merge away and home dataframes

```python
merge_games = home_teams.merge(away_teams, how='outer', left_index=True, right_index=True)
```

```python
list(merge_games.columns)
# merge_games = merge_games.drop(columns=['Away Match\xa0Up', 'Away MIN',])
merge_games['Home Wins'] = merge_games.apply(lambda x: 1 if x['Home W/L'] == "W" else 0, axis=1)
merge_games = merge_games.drop(columns=['Away Match\xa0Up', 'Away MIN', 'Home W/L', 'Away W/L'])
match_up = merge_games.pop('Home Match\xa0Up')
game_date = merge_games.pop('Home Game\xa0Date')
merge_games.insert(0, 'Date', game_date)
merge_games.insert(1, 'Game', match_up)
cleaned_games = merge_games
cleaned_games
```

```python
cleaned_games.to_csv('cleaned_games_stats.csv', mode='a', header=False)
```
